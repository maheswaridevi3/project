```python
df = completed_data.copy()
# Select numerical columns except AQI
columns_to_filter = [col for col in df.columns if df[col].dtype in ['float64', 'int64'] and col != "AQI"]

# Kalman filter function
def kalman_filter(data):
    n = len(data)
    x_est = np.nan_to_num(data[0])  # Initial estimate, replacing NaN with 0
    P = 1
    Q = 0.01  # Process noise covariance
    R = 0.1   # Measurement noise covariance
    filtered_data = np.zeros(n)

    for k in range(n):
        if np.isnan(data[k]):  # Skip missing values (NaN)
            filtered_data[k] = x_est
            continue

        # Prediction Step
        x_pred = x_est
        P_pred = P + Q

        # Kalman Gain Calculation
        K = P_pred / (P_pred + R)

        # Update Step (Correction)
        x_est = x_pred + K * (data[k] - x_pred)
        P = (1 - K) * P_pred

        # Store the filtered result
        filtered_data[k] = x_est

    return filtered_data

# Apply Kalman filter to selected columns
df_filtered = df.copy()
for col in columns_to_filter:
    df_filtered[col] = kalman_filter(df[col].values)

# Show the first few rows of the filtered data
df_filtered.head()
```

```python
# Select a few key pollutants for visualization
selected_columns = ["PM2.5","PM10","NO","NO2","NOx","NH3","CO","SO2","O3","Benzene","Toluene","Xylene"]

# Plot original vs. filtered data
fig, axes = plt.subplots(len(selected_columns), 1, figsize=(10, 20))
for i, col in enumerate(selected_columns):
    axes[i].plot(df[col], 'r--', alpha=0.6, label='Raw Data')
    axes[i].plot(df_filtered[col], 'b', label='Filtered Data', linewidth=1.5)
    axes[i].set_title(f"Kalman Filtering - {col}")
    axes[i].set_xlabel("Sample Index")
    axes[i].set_ylabel(col)
    axes[i].legend()
    axes[i].grid()

plt.tight_layout()
plt.show()
```
